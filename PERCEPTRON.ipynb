```python
import numpy as np
import matplotlib.pyplot as plt
```

# Data Set: Tabla de verdad Comperta AND

```python
inputs = numpy.array([[0,0],[0,1],[1,0],[1,1]])
outputs = numpy.array([0,0,0,1])
```

# Parámetros

```python
plt.figure(figsize=(3, 3))
plt.scatter(inputs[:,0],inputs[:,1], marker = "o", c=outputs, edgecolors= "k")
plt.xlabel('A')
plt.ylabel('B')
```

# Parámetros de entrenamiento:


*   Pesos W
*   bias
*   learning_rate



```python
weights = np.random.rand(2)
bias = np.random.rand(1)
learning_rate = 0.1
```

```python
bias
```

# Modelo

```python
# Definir una funcino para predecir
def predict(inputs):
  linear = np.dot(inputs, weights) + bias
  return np.where(linear >= 0, 1, 0)
```

# Predicción

```python
# ACTUALIZACION DE PESOS
def train(inputs, outputs, epoch):
  global weights, bias
  for epoch in range (epoch):
    total_error = 0
    for x, y in zip(inputs, outputs):
      y_predict = predict(x)
      error = y - y_predict
      total_error += abs(error)
      weights += learning_rate * error * x
      bias += learning_rate * error
    print(f"Epoch {epoch + 1}: Total Error = {total_error}")
```

```python
train(inputs, outputs, epoch = 10)
```

# Gráfica de límite de desición
image.png

```python
plt.figure(figsize=(3, 3))
plt.scatter(inputs[:,0],inputs[:,1], marker = "o", c=outputs, edgecolors= "k")
plt.xlabel('A')
plt.ylabel('B')

A = np.linspace(-0.5, 1.5, 100)
B = -(weights[0] * A + bias) / weights[1]
plt.plot(A, B, color='red')
```
